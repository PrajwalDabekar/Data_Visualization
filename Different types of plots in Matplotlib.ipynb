# Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
points = np.array([4,7,2,10,20])
```

```python
plt.plot(points)
plt.show()
```

```python
plt.plot(points,linestyle = 'dotted')
plt.show()
```

```python
plt.plot(points,linestyle = 'dashed')
plt.show()
```

# Bar Chart

```python
x = np.array(["X","y","Z","W"])
y = np.array([25,30,27,22])
```

```python
plt.bar(x,y)
plt.show()
```

# Scatter Plot

```python
x1 = np.array([20,15,11,7,8,10,25])
y1 = np.array([2,7,15,8,10,12,20])
```

```python
plt.scatter(x1,y1)
plt.show()
```

# Pie chart

```python
p = np.array([30,20,25,20])
plt.pie(p)
plt.show()
```
